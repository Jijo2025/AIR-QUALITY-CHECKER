```python
x=data[['CO','Ozone','NO2','SO2','PM2.5','PM10']]
y=data['AQI']
```

```python
plt.figure(figsize=(14, 7))

plt.plot(data['CO'], label='CO (mg/m3)')
plt.plot(data['Ozone'], label='Ozone (Âµg/m3)')
plt.plot(data['NO2'], label='NO2 (Âµg/m3)')
plt.plot(data['SO2'], label='SO2 (Âµg/m3)')
plt.plot(data['PM2.5'], label='PM2.5 (Âµg/m3)')
plt.plot(data['PM10'], label='PM10 (Âµg/m3)')

plt.title('Pollutant Levels Over Time')
plt.xlabel('Index')
plt.ylabel('Concentration')
plt.legend()
plt.grid(True)
plt.show()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y)
```

```python
from sklearn.linear_model import LinearRegression
reg=LinearRegression()
reg.fit(x_train,y_train)
y_pred=reg.predict(x_test)
```

```python

y_pred
```

```python
from sklearn.metrics import r2_score
r2_score(y_test,y_pred)
```

```python
from sklearn.metrics import mean_squared_error

print("RMSE = ",np.sqrt(mean_squared_error(y_test,y_pred)))
```

```python
from sklearn.neighbors import KNeighborsRegressor
```

```python
knnreg=KNeighborsRegressor()
knnreg.fit(x_train,y_train)
y_pred=knnreg.predict (x_test)
```

```python

y_pred
```

```python

r2_score(y_test,y_pred)
```

```python
print("RMSE = ",np.sqrt(mean_squared_error(y_test,y_pred)))
```

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
decireg=DecisionTreeRegressor()
decireg.fit(x_train,y_train)
y_pred=decireg.predict(x_test)
```

```python
y_pred
```

```python
r2_score(y_test,y_pred)
```

```python
print("RMSE = ",np.sqrt(mean_squared_error(y_test,y_pred)))
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
randreg=RandomForestRegressor()
randreg.fit(x_train,y_train)
y_pred=randreg.predict(x_test)
```
